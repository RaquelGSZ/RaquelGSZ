A.	 Deve-se implementar uma Lista Encadeada Simples em que: [EXIGÊNCIA DE CÓDIGO 1 de 7];
a.	O Nodo representa um cartão numerado contendo: número, cor e um ponteiro para o próximo;
b.	A lista é não circular, ou seja, seu último elemento aponta para nulo;


```python
# Nodo representando um cartão
class Nodo:
    def __init__(self, numero, cor):
        self.numero = numero      # Número do cartão
        self.cor = cor            # Cor do cartão
        self.proximo = None       # Ponteiro para o próximo nodo

# Lista Encadeada Simples (não circular)
class ListaEncadeadaSimples:
    def __init__(self):
        self.inicio = None  # Ponteiro para o primeiro nodo da lista
```

B.	 Deve-se implementar a função inserirSemPrioridade(nodo) em que: [EXIGÊNCIA DE CÓDIGO 2 de 7];
a.	 Deve-se andar pela lista a partir da cabeça (head) e inserir o nodo no final da lista.


```python
class ListaEncadeadaSimples:
    def __init__(self):
        self.inicio = None

    def inserirSemPrioridade(self, novo_nodo):
        if self.inicio is None:
            self.inicio = novo_nodo  # Lista vazia, novo nodo é o primeiro
        else:
            atual = self.inicio
            while atual.proximo is not None:
                atual = atual.proximo  # Anda até o último nodo
            atual.proximo = novo_nodo  # Insere no final
            # Criando alguns cartões (nodos)
cartao1 = Nodo(1, "vermelho")
cartao2 = Nodo(2, "azul")
cartao3 = Nodo(3, "verde")

# Criando a lista
lista = ListaEncadeadaSimples()

# Inserindo sem prioridade
lista.inserirSemPrioridade(cartao1)
lista.inserirSemPrioridade(cartao2)
lista.inserirSemPrioridade(cartao3)

```

C.	 Deve-se implementar a função inserirComPrioridade(nodo) em que: [EXIGÊNCIA DE CÓDIGO 3 de 7];
a.	 Deve-se andar pela lista a partir da cabeça (head) e inserir o nodo após todos os nodos com cor “A” que estão na lista.
b.	 O nodo inserido deve sempre estar antes de todos os nodos com cor “V”.


```python
class Nodo:
    def __init__(self, numero, cor):
        self.numero = numero
        self.cor = cor
        self.proximo = None

class ListaEncadeadaSimples:
    def __init__(self):
        self.inicio = None

    def inserirComPrioridade(self, novo_nodo):
        # Caso 1: Lista vazia
        if self.inicio is None:
            self.inicio = novo_nodo
            return

        anterior = None
        atual = self.inicio

        # Procurar onde inserir: após todos os "A", antes do primeiro "V"
        while atual is not None:
            if atual.cor == "A":
                anterior = atual  # Guardar o último nodo com cor "A"
            elif atual.cor == "V":
                break  # Parar antes do primeiro "V"
            elif anterior is None:
                anterior = None  # Nenhuma "A" antes, mas também não é "V"
            atual = atual.proximo

        if anterior is None:
            # Inserir no início (antes de todos, ou antes de "V")
            novo_nodo.proximo = self.inicio
            self.inicio = novo_nodo
        else:
            # Inserir após o último "A"
            novo_nodo.proximo = anterior.proximo
            anterior.proximo = novo_nodo
            # Criando cartões
cartao1 = Nodo(1, "A")
cartao2 = Nodo(2, "A")
cartao3 = Nodo(3, "V")
cartao4 = Nodo(4, "azul")  # Este será inserido com prioridade


# Criando lista
lista = ListaEncadeadaSimples()
lista.inserirSemPrioridade(cartao1)
lista.inserirSemPrioridade(cartao2)
lista.inserirSemPrioridade(cartao3)

# Inserir com prioridade
lista.inserirComPrioridade(cartao4)

# Função para imprimir a lista (ajuda na visualização)
def imprimir(lista):
    atual = lista.inicio
    while atual is not None:
        print(f"Número: {atual.numero}, Cor: {atual.cor}")
        atual = atual.proximo

# Visualizar resultado
imprimir(lista)
```

D.	 Deve-se implementar a função inserir() em que: [EXIGÊNCIA DE CÓDIGO 4 de 7];
a.	 Deve-se solicitar ao usuário a cor (“A” ou “V”).
b.	A partir da cor, o número (inteiro) do paciente deve ser atribuído automaticamente seguindo a ordem numérica. Por exemplo: o primeiro paciente “V” será o 1, o segundo 2, e assim por diante.
c.	 Deve-se criar um nodo com a cor e o número atribuído ao paciente.
d.	 Se a lista estiver vazia, a cabeça (head) da lista deve apontar para o nodo criado. Senão, se a cor do nodo for “V”, deve-se chamar a função inserirSemPrioridade(nodo). Senão, se a cor do nodo for “A”, deve-se chamar a função inserirComPriordade(nodo).


```python
class Nodo:
    def __init__(self, numero, cor):
        self.numero = numero
        self.cor = cor
        self.proximo = None


class ListaEncadeadaSimples:
    def __init__(self):
        self.inicio = None
        self.contadorA = 0
        self.contadorV = 0

    def inserirSemPrioridade(self, novo_nodo):
        if self.inicio is None:
            self.inicio = novo_nodo
        else:
            atual = self.inicio
            while atual.proximo is not None:
                atual = atual.proximo
            atual.proximo = novo_nodo

    def inserirComPrioridade(self, novo_nodo):
        if self.inicio is None:
            self.inicio = novo_nodo
            return

        anterior = None
        atual = self.inicio

        while atual is not None:
            if atual.cor == "A":
                anterior = atual
            elif atual.cor == "V":
                break
            atual = atual.proximo

        if anterior is None:
            novo_nodo.proximo = self.inicio
            self.inicio = novo_nodo
        else:
            novo_nodo.proximo = anterior.proximo
            anterior.proximo = novo_nodo

    def inserir(self):
        cor = input("Digite a cor do paciente (A para Amarelo, V para Vermelho): ").strip().upper()

        if cor not in ("A", "V"):
            print("Cor inválida. Use apenas 'A' ou 'V'.")
            return

        if cor == "A":
            self.contadorA += 1
            numero = self.contadorA
        else:  # cor == "V"
            self.contadorV += 1
            numero = self.contadorV

        novo_nodo = Nodo(numero, cor)

        if self.inicio is None:
            self.inicio = novo_nodo
        elif cor == "V":
            self.inserirSemPrioridade(novo_nodo)
        elif cor == "A":
            self.inserirComPrioridade(novo_nodo)


# Função para imprimir a lista (ajuda na visualização)
def imprimir(lista):
    atual = lista.inicio
    while atual:
        print(f"Número: {atual.numero}, Cor: {atual.cor}")
        atual = atual.proximo

lista = ListaEncadeadaSimples()

# Inserir 3 pacientes interativamente
for _ in range(3):
    lista.inserir()

# Imprimir o resultado
imprimir(lista)
```
